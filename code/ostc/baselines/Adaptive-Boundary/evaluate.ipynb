```python
import pandas as pd
final_df = []
for ratio in [0.75, 0.5, 0.25]:
    for dataset_name in ['atis', 'snips', 'linkedln', 'banking', 'clinc', 'mcid', 'reuters', 'stackoverflow', 'thucnews', 'news', 'ele']:
        for seed in range(5):
            try:
                file_path = f"results/{dataset_name}_{ratio}_results_{seed}.txt"
                df = pd.read_csv(file_path, header=None)
                df.columns = [i.split(":")[0].strip(' ') for i in df.iloc[1].tolist()]
                for col in df.columns:
                    df[col] = df[col].apply(lambda x: x.split(':')[1].strip(' '))
                df = df[['dataset', 'embedding', 'accuracy_all', 'f1_all', 'f1_ood', 'f1_id']]
                df.columns = ['dataset', 'method', 'ACC', 'F1', 'N-F1', 'K-F1']
                df['ratio'] = df['dataset'].apply(lambda x: x.split('_')[1])
                df['dataset'] = df['dataset'].apply(lambda x: x.split('_')[0])
                df['method'] = df['method'].apply(lambda x: 'AB-' + x)
                for metric_name in ['ACC', 'F1', 'N-F1', 'K-F1']:
                    sub_df = df[[ 'method', 'dataset', 'ratio'] + [metric_name]]
                    sub_df['value'] = sub_df[metric_name]
                    sub_df['metric'] = metric_name
                    sub_df = sub_df.drop(metric_name, axis=1)
                    sub_df['seed'] = seed
                    final_df.append(sub_df)
            except:
                print()
final_df = pd.concat(final_df).drop_duplicates(['method','dataset','metric', 'ratio', 'seed'])
final_df = final_df.sort_values(['method','dataset','metric', 'seed'])
final_df.to_csv('../../table_results/raw/AB_result.csv', index=None)
```
