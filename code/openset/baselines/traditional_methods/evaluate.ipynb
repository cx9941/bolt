```python
import pandas as pd
import os
final_df = []
for ratio in [0.75, 0.5, 0.25]:
    for dataset_name in ['atis', 'snips', 'clinc', 'stackoverflow', 'banking', 'mcid', 'reuters', 'thucnews' , 'linkedln', 'ele', 'news']:
    # for dataset_name in ['clinc']:
        for backbone in ['bert']:
            for seed in range(5):
                try:
                    file_path = f"outputs/{dataset_name}/{dataset_name}_{ratio}/{backbone}_{seed}/metrics/epoch_10_seed_{seed}.csv"
                    df = pd.read_csv(file_path, sep='\t')
                    df['dataset'] = dataset_name
                    df['method'] = df['Detector']
                    df['ratio'] = ratio
                    df = df[['dataset', 'method','accuracy', 'f1-score', 'N-f1', 'K-f1', 'ratio']]
                    df.columns = ['dataset', 'method', 'ACC', 'F1', 'N-F1', 'K-F1', 'ratio']
                    for metric_name in ['ACC', 'F1', 'N-F1', 'K-F1']:
                        sub_df = df[[ 'method', 'dataset', 'ratio'] + [metric_name]]
                        sub_df['value'] = sub_df[metric_name]
                        sub_df['metric'] = metric_name
                        sub_df = sub_df.drop(metric_name, axis=1)
                        sub_df['seed'] = seed
                        final_df.append(sub_df)
                    for ckpt in os.listdir(f"outputs/{dataset_name}/{dataset_name}_{ratio}/{backbone}_{seed}"):
                        if 'checkpoint' in ckpt:
                            print(f'rm -rf outputs/{dataset_name}/{dataset_name}_{ratio}/{backbone}_{seed}/{ckpt}')
                            os.system(f'rm -rf outputs/{dataset_name}/{dataset_name}_{ratio}/{backbone}_{seed}/{ckpt}')
                except:
                    print()
final_df = pd.concat(final_df).drop_duplicates(['method','dataset','metric', 'ratio', 'seed'])
final_df = final_df.sort_values(['method','dataset','metric', 'seed'])
final_df['method'] = final_df['method'].apply(lambda x: x.replace('OpenMax', '').replace('EnergyBased', 'OpenMax'))
final_df = final_df[final_df['method'].isin(['SHE', 'ViM', 'OpenMax'])]
final_df.to_csv('../../table_results/raw/tradition_result.csv', index=None)
final_df
```
