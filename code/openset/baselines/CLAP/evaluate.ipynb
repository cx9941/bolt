```python
import pandas as pd
import os
final_df = []
for ratio in [0.75, 0.5, 0.25]:
    for dataset_name in ['atis', 'snips', 'linkedln', 'banking', 'clinc', 'mcid', 'reuters', 'stackoverflow', 'thucnews', 'ele', 'news']:
        for seed in range(5):
            file_path = f"outputs/save_{dataset_name}_{ratio}_boundary_{seed}/results.csv"
            if os.path.exists(file_path):
                df = pd.read_csv(file_path)
                df = df[['dataset','known_cls_ratio','seed', 'Known','Open','F1-score','Accuracy']]
                df.columns = ['dataset','ratio','seed','K-F1', 'N-F1','F1', 'ACC']
                for metric_name in ['ACC', 'F1', 'N-F1', 'K-F1']:
                    sub_df = df[[ 'dataset', 'ratio'] + [metric_name]]
                    sub_df['value'] = sub_df[metric_name]
                    sub_df['metric'] = metric_name
                    sub_df = sub_df.drop(metric_name, axis=1)
                    sub_df['seed'] = seed
                    sub_df['method'] = 'CLAP'
                    final_df.append(sub_df)
            
                print(f"rm -rf outputs/save_{dataset_name}_{ratio}_model_{seed}")
                os.system(f"rm -rf outputs/save_{dataset_name}_{ratio}_model_{seed}")
final_df = pd.concat(final_df).drop_duplicates(['method','dataset','metric', 'ratio', 'seed'])
final_df = final_df.sort_values(['method','dataset','metric', 'seed'])
final_df.to_csv('../../table_results/raw/CLAP_result.csv', index=None)
```
