```python
import pandas as pd
import json
final_df = []
for ratio in [0.75, 0.5, 0.25]:
    for dataset_name in ['clinc', 'stackoverflow', 'banking', 'atis', 'snips', 'mcid', 'reuters', 'thucnews' , 'linkedln', 'ele', 'news']:
        for backbone in ['bert']:
            for seed in range(5):
                try:
                    file_path = f"metrics/{dataset_name}_{int(ratio*100)}/no_lof_{seed}.json"
                    df = pd.DataFrame([json.load(open(file_path, 'r'))])
                    df['dataset'] = dataset_name
                    df['method'] = 'SCL'
                    df['ratio'] = ratio
                    for metric_name in ['ACC', 'F1', 'N-F1', 'K-F1']:
                        sub_df = df[[ 'method', 'dataset', 'ratio'] + [metric_name]]
                        sub_df['value'] = sub_df[metric_name]
                        sub_df['metric'] = metric_name
                        sub_df = sub_df.drop(metric_name, axis=1)
                        sub_df['seed'] = seed
                        final_df.append(sub_df)
                except:
                    print()
final_df = pd.concat(final_df).drop_duplicates(['method','dataset','metric', 'ratio', 'seed'])
final_df = final_df.sort_values(['method','dataset','metric', 'seed'])
final_df['value'] = final_df['value']*100
final_df.to_csv('../../table_results/raw/SCL_result.csv', index=None)
final_df
```
