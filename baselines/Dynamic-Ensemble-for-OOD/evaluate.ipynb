```python
import pandas as pd
import os
final_df = []
for ratio in [0.25, 0.5, 0.75]:
    for dataset_name in ['atis', 'snips', 'clinc', 'stackoverflow', 'banking', 'mcid', 'reuters', 'thucnews' , 'linkedln', 'ele', 'news']:
        for seed in range(5):
            tag = True
            for eval_mode in ['knn', 'lof_cosine', 'energy', 'entropy', 'lof_euclidean', 'maha', 'maxLogit', 'msp', 'odin']:
                file_path = f"model_output/{dataset_name}_{ratio}/ad30.10.2_lr2.0e-05__epoch50__lossce_and_div_drop-last-layer__batchsize64__lambda0.1__scale1.51.2/{eval_mode}_{dataset_name}_{ratio}_{seed}.csv"
                if os.path.exists(file_path):
                    df = pd.read_csv(file_path)
                    df = df[df['strategy']=='pabee']
                    df = df[['ACC-all', 'F1', 'F1-ood', 'F1-ind']]
                    df.columns = ['ACC', 'F1','N-F1', 'K-F1']
                    for metric_name in ['ACC', 'F1', 'N-F1', 'K-F1']:
                        sub_df = df[[metric_name]]
                        sub_df['value'] = sub_df[metric_name]
                        sub_df['metric'] = metric_name
                        sub_df = sub_df.drop(metric_name, axis=1)
                        sub_df['seed'] = seed
                        sub_df['method'] = f'DyEn-{eval_mode}'
                        sub_df['dataset'] = dataset_name
                        sub_df['ratio'] = ratio
                        final_df.append(sub_df)
                else:
                    tag = False
            if tag:
                print(f"rm -rf model_output/{dataset_name}_{ratio}/ad30.10.2_lr2.0e-05__epoch50__lossce_and_div_drop-last-layer__batchsize64__lambda0.1__scale1.51.2/_{dataset_name}_{ratio}_{seed}.pt")
                os.system(f"rm -rf model_output/{dataset_name}_{ratio}/ad30.10.2_lr2.0e-05__epoch50__lossce_and_div_drop-last-layer__batchsize64__lambda0.1__scale1.51.2/_{dataset_name}_{ratio}_{seed}.pt")
final_df = pd.concat(final_df).drop_duplicates(['method','dataset','metric', 'ratio', 'seed'])
final_df = final_df.sort_values(['method','dataset','metric', 'seed'])
final_df.to_csv('../../table_results/raw/DyEn_result.csv', index=None)
final_df
```
